```python
import os

# Directorio raz que contiene los subdirectorios organizados por clase
root_dir = "/home/andres/ambientesVirtuales/PruebasML/Practicas/DatosEntrenamiento/correccion"

# Diccionario que asigna nombres de directorios a ndices de clases
class_mapping = {
    "doctor": 0,
    "mecanico": 1,
    "domestico": 2,
    "salvaje": 3,
    "noche": 4,
    "dia": 5,
    "montaa": 6,
    "playa": 7,
    "granja": 8,
    "dulce": 9,
    "saludable": 10
}

# Recorre cada subdirectorio
for class_name, class_index in class_mapping.items():
    class_dir = os.path.join(root_dir, class_name)

    if not os.path.isdir(class_dir):
        print(f"El directorio {class_dir} no existe. Lo omito.")
        continue

    # Recorre los archivos en el directorio de la clase
    for filename in os.listdir(class_dir):
        if filename.endswith(".txt"):
            file_path = os.path.join(class_dir, filename)

            # Leer el archivo .txt y actualizar la clase
            with open(file_path, "r") as file:
                lines = file.readlines()

            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) == 5:  # Asegrate de que tenga el formato esperado
                    parts[0] = str(class_index)  # Actualiza la clase
                    updated_lines.append(" ".join(parts))

            # Escribir las lneas actualizadas al archivo .txt
            with open(file_path, "w") as file:
                file.write("\n".join(updated_lines))
            
            print(f"Archivo actualizado: {file_path}")

print("Proceso completado.")
```
